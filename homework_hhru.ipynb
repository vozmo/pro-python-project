1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
def extract_education_level(raw_string):
    """
    Функция для извлечения уровня образования из строки.

    return:
    str or None: Уровень образования или None, если уровень не определен.
    """
    words = " ".join(raw_string.split()[:2]).lower()
    categories = ["неоконченное высшее", "среднее специальное", "среднее", "высшее"]
    for category in categories:
        if category in words:
            return category
    return None

# Применяем функцию для извлечения уровня образования и создаем новый столбец
df["Образование"] = df["Образование и ВУЗ"].apply(extract_education_level)

# Выводим количество значений каждого уровня образования
print(df["Образование"].value_counts())

# Удаляем столбец "Образование и ВУЗ", так как он больше не нужен
df.drop("Образование и ВУЗ", axis=1, inplace=True)
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def process_experience(value):
    """Обрабатывает строку с опытом работы."""
    if pd.isna(value) or value == "Не указано":
        return np.nan, False
    split_str = value.split()[2:6]
    if split_str[-1] not in ["месяца", "месяцев"]:
        split_str = split_str[:2]
    return split_str, True

def convert_to_months(split_lst_flag):
    """Преобразует данные опыта работы в месяцы."""
    split_lst, flag = split_lst_flag
    if not flag:
        return np.nan

    total_months = 0
    if len(split_lst) == 4:
        split_lst = [[split_lst[0], split_lst[1]], [split_lst[2], split_lst[3]]]
    else:
        split_lst = [split_lst]
    for num, type_date in split_lst:
        num = int(num)
        if type_date in ["лет", "года", "год"]:
            total_months += num * 12
        else:
            total_months += num
    return total_months

def get_exp(value):
    """Возвращает количество месяцев опыта работы."""
    return convert_to_months(process_experience(value))

df["Опыт работы (месяц)"] = df["Опыт работы"].apply(get_exp)
df.drop("Опыт работы", axis=1, inplace=True)

median_experience = df['Опыт работы (месяц)'].median()
print(f"Медианное значение опыта работы (в месяцах): {median_experience}")
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def extract_town(raw_string):
    """
    Извлекает название города из строки с информацией о городе, переезде и командировках.

    Параметры:
    raw_string (str): Строка с информацией о городе, переезде и командировках.

    Возвращает:
    str: Название города.
    """
    # Извлекаем первое слово из строки и удаляем лишние пробелы
    town_name = raw_string.split(",")[0].strip()

    # Проверяем, является ли город Москвой или Санкт-Петербургом
    if town_name in ["Москва", "Санкт-Петербург"]:
        return town_name

    # Проверяем, входит ли город в список крупных городов
    big_towns = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград']
    if town_name in big_towns:
        return "город-миллионник"

    # Возвращаем "другие" для всех остальных случаев
    return "другие"

def extract_relocation_preference(raw_string):
    """
    Определяет готовность к переезду на основе строки с информацией о городе, переезде и командировках.

    Параметры:
    raw_string (str): Строка с информацией о городе, переезде и командировках.

    Возвращает:
    bool: Флаг готовности к переезду (True - готов, False - не готов).
    """
    # Извлекаем подстроку, содержащую информацию о переезде
    string_with_relocate = [i for i in raw_string.split(",") if "перее" in i][0]
    # Проверяем, отсутствует ли в подстроке отрицание
    return "не" not in string_with_relocate

def extract_business_trip_preference(raw_string):
    """
    Определяет готовность к командировкам на основе строки с информацией о городе, переезде и командировках.

    Параметры:
    raw_string (str): Строка с информацией о городе, переезде и командировках.

    Возвращает:
    bool: Флаг готовности к командировкам (True - готов, False - не готов).
    """
    # Извлекаем подстроку, содержащую информацию о командировках
    string_with_relocate = [i for i in raw_string.split(",") if "командировк" in i]
    # Если подстрока не найдена, возвращаем False
    if len(string_with_relocate) == 0:
        return False
    # Проверяем, отсутствует ли в подстроке отрицание
    return "не" not in string_with_relocate[0]

# Применяем функции для извлечения информации о городе, переезде и командировках
df["Город"] = df["Город, переезд, командировки"].apply(extract_town)
df["Готовность к переезду"] = df["Город, переезд, командировки"].apply(extract_relocation_preference)
df["Готовность к командировкам"] = df["Город, переезд, командировки"].apply(extract_business_trip_preference)

# Удаляем столбец "Город, переезд, командировки", так как он больше не нужен
df.drop("Город, переезд, командировки", axis=1, inplace=True)
```

```python
# Рассчитываем процент проживающих в Санкт-Петербурге
percentage_in_spb = (df["Город"] == "Санкт-Петербург").mean() * 100
print(f"Процент проживающих в Санкт-Петербурге: {percentage_in_spb:.2f}%")

# Рассчитываем процент готовых на переезд и командировки
relocation_and_business_trip_percentage = ((df["Готовность к командировкам"] & df["Готовность к переезду"]).mean()) * 100
print(f"Процент готовых на переезд и командировки: {relocation_and_business_trip_percentage:.2f}%")
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
def get_one_hot_encoding(column):
    """
    Создает кодировку One-Hot для столбца DataFrame.

    Параметры:
    column (pandas.Series): Столбец DataFrame, для которого требуется создать кодировку One-Hot.

    Возвращает:
    pandas.DataFrame: DataFrame с кодировкой One-Hot.
    """
    # Получаем уникальные значения из столбца
    all_values = set()
    for row in column:
        all_values.update(row.split(", "))
    all_values = list(all_values)

    # Создаем словарь для хранения кодировки One-Hot
    dct_ohe = {i: [] for i in all_values}

    # Заполняем словарь значениями кодировки One-Hot
    for row in column:
        for key in all_values:
            dct_ohe[key].append(int(key in row))

    return pd.DataFrame(dct_ohe).set_index(column.index)

# Создаем кодировку One-Hot для столбцов "График" и "Занятость"
ohe_timetable = get_one_hot_encoding(df["График"])
ohe_busy = get_one_hot_encoding(df["Занятость"])

# Объединяем DataFrame с исходным DataFrame
df = pd.concat([df, ohe_timetable, ohe_busy], axis=1)

# Удаляем столбцы "График" и "Занятость"
df.drop(["Занятость", "График"], axis=1, inplace=True)

# Рассчитываем количество строк, где присутствуют и "проектная работа", и "волонтерство"
project_and_volunteer_count = np.logical_and(df["проектная работа"], df["волонтерство"]).sum()
print(f"Количество строк с 'проектная работа' и 'волонтерство': {project_and_volunteer_count}")

# Рассчитываем количество строк, где присутствуют и "вахтовый метод", и "гибкий график"
shift_and_flexible_count = np.logical_and(df["вахтовый метод"], df["гибкий график"]).sum()
print(f"Количество строк с 'вахтовый метод' и 'гибкий график': {shift_and_flexible_count}")

```

* "proportion" - пропорция, 
* "close" - цена закрытия (последний зафиксированный курс валюты на указанный день).


Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Пропорция - это число, за сколько единиц валюты указан курс в таблице с курсами. Например, для казахстанского тенге курс на 20.08.2019 составляет 17.197 руб. за 100 тенге, тогда итоговый курс равен - 17.197 / 100 = 0.17197 руб за 1 тенге.
Воспользуйтесь этой информацией в ваших преобразованиях.

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Загружаем данные о курсах валют из файла ExchangeRates.csv, выбираем только нужные столбцы
df_currency = pd.read_csv("ExchangeRates.csv")[["currency", "date", "proportion", "close"]]

# Вычисляем реальное значение валюты и создаем новый столбец "real_currency"
df_currency["real_currency"] = df_currency["close"] / df_currency["proportion"]

# Устанавливаем индекс DataFrame на основе столбцов "currency" и "date"
df_currency = df_currency.set_index(["currency", "date"])
```

```python
from datetime import datetime

def get_payment_in_rubles(row):
    """
    Преобразует значение ЗП в рубли на основе курса валют.

    Параметры:
    row (pandas.Series): Значение из таблицы, которое необходимо перевести в рубли

    Возвращает:
    float: Значение ЗП в рублях.
    """

    # Разделяем значение ЗП на числовую часть и валюту
    num, currency = row["ЗП"].split()
    num = float(num)

    # Если валюта уже в рублях, возвращаем значение ЗП
    if currency == "руб.":
        return num

    # Преобразуем названия валют в соответствующие коды
    dct_transform_cur_name = {"бел.руб.":"BYN", "сум":"UZS", "грн.": "UAH"}
    if currency in dct_transform_cur_name.keys():
        currency = dct_transform_cur_name[currency]

    # Преобразуем дату в нужный формат
    date = datetime.strptime(row["Обновление резюме"], "%d.%m.%Y %H:%M").strftime("%d/%m/%y")

    # Получаем курс валюты на указанную дату и преобразуем значение ЗП в рубли
    return df_currency.loc[currency, date].real_currency * num

# Применяем функцию для преобразования ЗП в рубли
df["ЗП (руб)"] = df.apply(get_payment_in_rubles, axis=1)

# Удаляем столбец "ЗП"
df.drop("ЗП", axis=1, inplace=True)
```

```python
import numpy as np
import seaborn as sns

def plot_age_distribution(data):
    """
    Функция для построения гистограммы распределения возраста

    :param data: Входные данные (DataFrame)
    :return: None
    """
    # Применение логарифмического преобразования к возрасту
    log_age = np.log(data['Возраст'])

    # Построение гистограммы
    histplot = sns.histplot(log_age, kde=False)

    # Вычисление среднего и стандартного отклонения
    mean = log_age.mean()
    std = log_age.std()

    # Добавление вертикальных линий
    histplot.axvline(mean, color='k', lw=2, label='Среднее')  # Среднее
    histplot.axvline(mean - 3*std, color='r', lw=2, ls='--', label='-3 ст. отклонения')  # -3 сигмы
    histplot.axvline(mean + 3*std, color='r', lw=2, ls='--', label='+3 ст. отклонения')  # +3 сигмы

    # Добавление легенды
    histplot.legend()

# Построение гистограммы распределения возраста
plot_age_distribution(df)
```

```python
import numpy as np

def filter_outliers(data):
    # Вычисление z-оценок
    z_scores = (data - data.mean()) / data.std()

    # Создание маски для выбросов
    mask = ((z_scores > 4) | (z_scores < -3))

    return mask

mask_filter = filter_outliers(df["Возраст"])
```

```python
df = df[~mask_filter]
```
